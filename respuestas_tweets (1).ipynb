```python
import pandas as pd
import openai

url2 = 'https://raw.githubusercontent.com/kevingonzalezd/csv/main/filtered_clave.csv'
df2 = pd.read_csv(url2)

df2= df2[df2['Clasificacion']=='rendimiento']

llave='lallave'
df_reducido = df2.head(3)


key = llave
openai.api_key = key
modelo = 'gpt-4-turbo'

system =' Responde de forma breve ingeniosa graciosa para hacer reir a los usuarios, solo da la respuesta'

user = """


Revisa el humor de los tweets de las empresas en el siguiente link:

https://raw.githubusercontent.com/kevingonzalezd/csv/main/training_data.csv

habla como si estuvieras escribiendo en whatsapp de formma informal

Tu tarea es ayudarme a formular una respuesta que sea consistente con el estilo humor√≠stico observado en los ejemplos proporcionados.
No uses signos de admiraci√≥n (!¬°) que sea informal la respuesta, no uses emojis

Bas√°ndome en este estilo, necesito crear respuestas graciosas y divertidas a tweets de clientes.
 A continuaci√≥n, se presenta un tweet de un cliente al que debo responder usando el mismo tono humor√≠stico
 y creativo que caracteriza a las empresas:

"""


def generar_respuesta(tweet):
    mensajes = [
        {'role': 'system', 'content':system},
        {'role': 'user', 'content': user + tweet}
    ]

    respuesta = openai.ChatCompletion.create(
        model=modelo,
        messages=mensajes,
        temperature=0.8,
        max_tokens=1000
    )

    return respuesta["choices"][0]["message"]["content"]

# Aplicar la funci√≥n a cada tweet en el DataFrame reducido
df_reducido['respuesta '] = df_reducido['tweet'].apply(generar_respuesta)


df_reducido
```
