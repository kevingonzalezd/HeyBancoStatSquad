### **Exploracion Columnas con Emojis para confirmar que no se contradiga su significado en su mayoria**

```python
import pandas as pd
import re

# Suponemos que tu DataFrame se llama df y est√°s interesado en revisar la columna 'tweet'
def count_emojis(text):
    # Esta es una expresi√≥n regular que cubre una gama amplia de emojis
    emoji_pattern = re.compile("["
                               "\U0001F600-\U0001F64F"  # emoticons
                               "]+", flags=re.UNICODE)
    return len(emoji_pattern.findall(text))

# Aplicamos la funci√≥n count_emojis a cada elemento en la columna 'tweet'
df_horas['emoji_count'] = df_horas['tweet'].apply(count_emojis)

# Sumamos todos los valores de 'emoji_count' para obtener el n√∫mero total de emojis en la columna
total_emojis = df_horas['emoji_count'].sum()
print(f"Total de emojis en la columna 'tweet': {total_emojis}")
```

### **eliminacion acentos lista de palabras**

```python
import pandas as pd
import unidecode

# Aseg√∫rate de que df es tu DataFrame

# Funci√≥n para eliminar acentos de una lista de palabras
def remove_accents_from_list(words_list):
    return [unidecode.unidecode(word) for word in words_list]

# Aplicar la funci√≥n para eliminar acentos en la columna 'tokens'
df['tokens'] = df['tokens'].apply(remove_accents_from_list)

df
```

# **Primera Prueba Analisis Sentimiento (no escogido bajo rendimiento) Textblob**

```python
import pandas as pd
from textblob import TextBlob

# Suponiendo que ya tienes tu DataFrame cargado como df y que se ve como el mostrado

# Convertir la lista de tokens a una cadena de texto para an√°lisis
df['text_clean'] = df_tokens['tokens'].apply(lambda tokens: ' '.join(tokens))

# Crear una funci√≥n para analizar el sentimiento
def analizar_sentimiento(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positivo'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negativo'

# Aplicar la funci√≥n de an√°lisis de sentimiento a la columna con el texto limpio
df['sentimiento'] = df['text_clean'].apply(analizar_sentimiento)

# Mostrar el DataFrame con la columna de sentimiento a√±adida
#print(df[['tokens', 'text_clean', 'sentimiento']])

df
```

# **Clasificacion por medio de jerarquias en base a menor cantidad de tweets con la palabra mayor especidicidad**

```python
import pandas as pd


# Supongamos que tu DataFrame se llama df

# Lista de palabras clave en orden jer√°rquico
keywords = ['cashback', 'rendimiento', 'servicio', 'tarjeta', 'banco','app', 'credito', 'inversion','recomiendo','recompensa']

# Funci√≥n para determinar la clasificaci√≥n basada en palabras clave
def classify_text(text):
    for keyword in keywords:
        if keyword in text.lower():
            return keyword
    return "Otro"  # Para textos que no contienen ninguna de las palabras clave

# Aplicar la funci√≥n para crear la nueva columna
filtered_clave['Clasificacion'] = filtered_clave['text_clean'].apply(classify_text)

filtered_clave
```

# **Creacion Diccionario de Hastags**

```python
keywords_to_hashtags = {
    'cashback': ['#HeyBancoReembolso', '#HeyBancoAhorro', '#HeyBancoPremios', '#HeyBancoCompras', '#HeyBancoOfertas'],
    'rendimiento': ['#HeyBancoRendimiento', '#HeyBancoEficiencia', '#HeyBancoAltoRendimiento', '#HeyBancoInversiones', '#HeyBancoROI'],
    'servicio': ['#HeyBancoAtenci√≥nAlCliente', '#HeyBancoSoporte', '#HeyBancoServicioT√©cnico', '#HeyBancoCuidadoCliente', '#HeyBancoCalidadServicio'],
    'tarjeta': ['#HeyBancoTarjetaDeCr√©dito', '#HeyBancoBanca', '#HeyBancoFinanzas', '#HeyBancoTarjetaDeD√©bito', '#HeyBancoPago'],
    'banco': ['#HeyBancoBanco', '#HeyBancoFinanzas', '#HeyBancoDinero', '#HeyBancoAhorros', '#HeyBancoInversi√≥n'],
    'app': ['#HeyBancoAplicaci√≥n', '#HeyBancoTecnolog√≠a', '#HeyBancoAppM√≥vil', '#HeyBancoInnovaci√≥n', '#HeyBancoApps'],
    'credito': ['#HeyBancoCr√©dito', '#HeyBancoFinanzas', '#HeyBancoPr√©stamo', '#HeyBancoPr√©stamoBancario', '#HeyBancoCalificaci√≥nCrediticia'],
    'inversion': ['#HeyBancoInversi√≥n', '#HeyBancoAcciones', '#HeyBancoPatrimonio', '#HeyBancoFinanzas', '#HeyBancoInversor'],
    'recomiendo': ['#HeyBancoRecomiendo', '#HeyBancoConsejo', '#HeyBancoSugerencia', '#HeyBancoMejorOpci√≥n', '#HeyBancoDebesProbar'],
    'recompensa': ['#HeyBancoRecompensa', '#HeyBancoBono', '#HeyBancoPremio', '#HeyBancoRegalo', '#HeyBancoGanancia']
}
```

```python
def generate_hashtags_from_classification(classification):
    return keywords_to_hashtags.get(classification.lower(), [])
```

```python
# Asegurarse de que 'filtered_clave' es el DataFrame principal y no una vista/copi
filtered_clave['hashtags'] = filtered_clave['Clasificacion'].apply(generate_hashtags_from_classification)

# Mostrar el DataFrame resultante
filtered_clave

```

```python
import openai

# Configura tu clave API de OpenAI (aseg√∫rate de manejar esto de forma segura, por ejemplo, usando variables de entorno)
openai.api_key = 'introducir_llave'

def generate_insights(text):
    response = openai.ChatCompletion.create(
        model="gpt-4-turbo",  # Aseg√∫rate de que es el modelo adecuado
        messages=[
            {"role": "system", "content": "You are an analytical assistant."},
            {"role": "user", "content": f"Analiza este comentario de un usuario de Hey Banco y proporcioname insights relevantes para mejorar como empresa: {text}"}
        ]
    )
    return response['choices'][0]['message']['content'].strip()
```
